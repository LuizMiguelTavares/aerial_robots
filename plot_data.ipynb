```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
def load_csv(file_path):
    try:
        data = pd.read_csv(file_path)
        print(f"Successfully loaded data from {file_path}")
        return data
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return None

# Plot desired vs actual positions
def plot_positions(data):
    if data is None:
        return

    # Extract relevant data
    time = np.array(data['Time'])
    pos_x = np.array(data['Pos_X'])
    pos_y = np.array(data['Pos_Y'])
    pos_z = np.array(data['Pos_Z'])
    pos_x_des = np.array(data['Pos_X_des'])
    pos_y_des = np.array(data['Pos_Y_des'])
    pos_z_des = np.array(data['Pos_z_des'])
    
    # Create subplots for X, Y, Z positions
    plt.figure(figsize=(12, 8))

    # X position
    plt.subplot(3, 1, 1)
    plt.plot(time, pos_x, label="Actual X")
    plt.plot(time, pos_x_des, label="Desired X", linestyle="--")
    plt.xlabel("Time (s)")
    plt.ylabel("Position X (m)")
    plt.legend()
    plt.grid()

    # Y position
    plt.subplot(3, 1, 2)
    plt.plot(time, pos_y, label="Actual Y")
    plt.plot(time, pos_y_des, label="Desired Y", linestyle="--")
    plt.xlabel("Time (s)")
    plt.ylabel("Position Y (m)")
    plt.legend()
    plt.grid()

    # Z position
    plt.subplot(3, 1, 3)
    plt.plot(time, pos_z, label="Actual Z")
    plt.plot(time, pos_z_des, label="Desired Z", linestyle="--")
    plt.xlabel("Time (s)")
    plt.ylabel("Position Z (m)")
    plt.legend()
    plt.grid()

    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()

def plot_velocities(data):
    if data is None:
        return

    # Extract relevant data
    time = np.array(data['Time'])
    vel_x = np.array(data['Vel_X'])
    vel_y = np.array(data['Vel_Y'])
    vel_z = np.array(data['Vel_Z'])
    vel_x_des = np.array(data['Vel_X_des'])
    vel_y_des = np.array(data['Vel_Y_des'])
    vel_z_des = np.array(data['Vel_Z_des'])
    
    # Create subplots for X, Y, Z velocities
    plt.figure(figsize=(12, 8))

    # X velocity
    plt.subplot(3, 1, 1)
    plt.plot(time, vel_x, label="Velocity X")
    plt.plot(time, vel_x_des, label="Desired Velocity X", linestyle="--")
    plt.xlabel("Time (s)")
    plt.ylabel("Velocity X (m/s)")
    plt.legend()
    plt.grid()

    # Y velocity
    plt.subplot(3, 1, 2)
    plt.plot(time, vel_y, label="Velocity Y")
    plt.plot(time, vel_y_des, label="Desired Velocity Y", linestyle="--")
    plt.xlabel("Time (s)")
    plt.ylabel("Velocity Y (m/s)")
    plt.legend()
    plt.grid()

    # Z velocity
    plt.subplot(3, 1, 3)
    plt.plot(time, vel_z, label="Velocity Z")
    plt.plot(time, vel_z_des, label="Desired Velocity Z", linestyle="--")
    plt.xlabel("Time (s)")
    plt.ylabel("Velocity Z (m/s)")
    plt.legend()
    plt.grid()

    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()
    
def plot_position_error(data):
    if data is None:
        return

    # Extract relevant data
    time = np.array(data['Time'])
    pos_x = np.array(data['Pos_X'])
    pos_x_des = np.array(data['Pos_X_des'])
    pos_y = np.array(data['Pos_Y'])
    pos_y_des = np.array(data['Pos_Y_des'])
    pos_z = np.array(data['Pos_Z'])
    pos_z_des = np.array(data['Pos_z_des'])
    
    # Calculate position errors
    pos_x_error = pos_x - pos_x_des
    pos_y_error = pos_y - pos_y_des
    pos_z_error = pos_z - pos_z_des
    
    # Create subplots for X, Y, Z positions
    plt.figure(figsize=(12, 8))

    # X position
    plt.subplot(3, 1, 1)
    plt.plot(time, pos_x_error, label="Position X Error")
    plt.ylim(-0.1, 0.1)
    plt.xlabel("Time (s)")
    plt.ylabel("Position X Error (m)")
    plt.legend()
    plt.grid()

    # Y position
    plt.subplot(3, 1, 2)
    plt.plot(time, pos_y_error, label="Position Y Error")
    plt.ylim(-0.1, 0.1)
    plt.xlabel("Time (s)")
    plt.ylabel("Position Y Error (m)")
    plt.legend()
    plt.grid()

    # Z position
    plt.subplot(3, 1, 3)
    plt.plot(time, pos_z_error, label="Position Z Error")
    plt.ylim(-0.1, 0.1)
    plt.xlabel("Time (s)")
    plt.ylabel("Position Z Error (m)")
    plt.legend()
    plt.grid()

    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()

def plot_velocity_error(data):
    if data is None:
        return

    # Extract relevant data
    time = np.array(data['Time'])
    vel_x = np.array(data['Vel_X'])
    vel_x_des = np.array(data['Vel_X_des'])
    vel_y = np.array(data['Vel_Y'])
    vel_y_des = np.array(data['Vel_Y_des'])
    vel_z = np.array(data['Vel_Z'])
    vel_z_des = np.array(data['Vel_Z_des'])
    
    # Calculate velocity errors
    vel_x_error = vel_x - vel_x_des
    vel_y_error = vel_y - vel_y_des
    vel_z_error = vel_z - vel_z_des
    
    # Create subplots for X, Y, Z velocities
    plt.figure(figsize=(12, 8))

    # X velocity
    plt.subplot(3, 1, 1)
    plt.plot(time, vel_x_error, label="Velocity X Error")
    plt.xlabel("Time (s)")
    plt.ylabel("Velocity X Error (m/s)")
    plt.ylim(-0.1, 0.1)
    plt.legend()
    plt.grid()

    # Y velocity
    plt.subplot(3, 1, 2)
    plt.plot(time, vel_y_error, label="Velocity Y Error")
    plt.xlabel("Time (s)")
    plt.ylabel("Velocity Y Error (m/s)")
    plt.ylim(-0.1, 0.1)
    plt.legend()
    plt.grid()

    # Z velocity
    plt.subplot(3, 1, 3)
    plt.plot(time, vel_z_error, label="Velocity Z Error")
    plt.xlabel("Time (s)")
    plt.ylabel("Velocity Z Error (m/s)")
    plt.ylim(-0.1, 0.1)
    plt.legend()
    plt.grid()

    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()
```

```python
x_pos = np.array(data['Pos_X'])
y_pos = np.array(data['Pos_Y'])
z_pos = np.array(data['Pos_Z'])

time = np.array(data['Time'])

positions = np.array([x_pos, y_pos, z_pos]).T
positions
```

```python
# Derivation to find velocity
vel_no_filter = []
vel_first_order_filter = []
alpha_first_order = 0.1

vel_second_order_filter = []
alpha_second_order = 0.3

for i, position in enumerate(positions):
    if i == 0:
        # Vel no filter
        vel_no_filter.append(np.zeros(3))
        
        # Vel first order filter
        vel_first_order_filter.append(np.zeros(3))
        last_vel_first_order = np.zeros(3)
        
        # Vel second order filter
        vel_second_order_filter.append(np.zeros(3))
        last_vel = np.zeros(3)
        last_last_vel = np.zeros(3)
        continue
    
    # Instantaneous velocity
    inst_vel = (position - positions[i-1]) / (time[i] - time[i-1])
    
    # No filter
    vel_no_filter.append(inst_vel)
    
    # First order filter
    vel_filt_first_order = alpha_first_order * inst_vel + (1 - alpha_first_order) * last_vel_first_order   
    vel_first_order_filter.append(vel_filt_first_order)
    last_vel_first_order = vel_filt_first_order
    
    # Second order filter 
    vel_filt_second_order = alpha_second_order * alpha_second_order * inst_vel + alpha_second_order * (1 - alpha_second_order) * last_vel + (1 - alpha_second_order) * last_last_vel
    vel_second_order_filter.append(vel_filt_second_order)
    last_last_vel = last_vel
    last_vel = vel_filt_second_order
```

```python
# plot velocities
vel_no_filter = np.array(vel_no_filter)
vel_1_order = np.array(vel_first_order_filter)
vel_2_order = np.array(vel_second_order_filter)

plt.figure(figsize=(12, 8))

# X velocity
plt.subplot(3, 1, 1)
plt.plot(time, vel_no_filter[:, 0], label="Unfiltered Velocity X")
plt.plot(time, vel_1_order[:, 0], label="Filtered Velocity X")
plt.plot(time, vel_2_order[:, 0], label="Filtered Velocity X 2nd Order")
plt.xlabel("Time (s)")
plt.ylabel("Velocity X (m/s)")
plt.legend()
plt.grid()

# Y velocity
plt.subplot(3, 1, 2)
plt.plot(time, vel_no_filter[:, 1], label="Unfiltered Velocity Y")
plt.plot(time, vel_1_order[:, 1], label="Filtered Velocity Y")
plt.plot(time, vel_2_order[:, 1], label="Filtered Velocity Y 2nd Order")
plt.xlabel("Time (s)")
plt.ylabel("Velocity Y (m/s)")
plt.legend()
plt.grid()

# Z velocity
plt.subplot(3, 1, 3)
plt.plot(time, vel_no_filter[:, 2], label="Unfiltered Velocity Z")
plt.plot(time, vel_1_order[:, 2], label="Filtered Velocity Z")
plt.plot(time, vel_2_order[:, 2], label="Filtered Velocity Z 2nd Order")
plt.xlabel("Time (s)")
plt.ylabel("Velocity Z (m/s)")
plt.legend()
plt.grid()
```

```python
# Apply filt filt on the "vel_no_filter" data

from scipy.signal import butter, filtfilt

def butter_lowpass(cutoff, fs, order=5):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = filtfilt(b, a, data)
    return y

# Filter requirements.
order = 6
fs = 30.0       # sample rate, Hz
cutoff = 0.9  # desired cutoff frequency of the filter, Hz

# Get the filter coefficients so we can check its frequency response.
b, a = butter_lowpass(cutoff, fs, order)

# Apply the filter to the data
# Filter each velocity component separately
vel_no_filter_filtered = np.array([
    butter_lowpass_filter(vel_no_filter[:, i], cutoff, fs, order)
    for i in range(vel_no_filter.shape[1])
]).T

plt.figure(figsize=(12, 8))

# X velocity
plt.subplot(3, 1, 1)
plt.plot(time, vel_no_filter[:, 0], label="Unfiltered Velocity X")
plt.plot(time, vel_no_filter_filtered[:, 0], label="Filtered Velocity X")
plt.xlabel("Time (s)")
plt.ylabel("Velocity X (m/s)")
plt.legend()
plt.grid()

# Y velocity
plt.subplot(3, 1, 2)
plt.plot(time, vel_no_filter[:, 1], label="Unfiltered Velocity Y")
plt.plot(time, vel_no_filter_filtered[:, 1], label="Filtered Velocity Y")
plt.xlabel("Time (s)")
plt.ylabel("Velocity Y (m/s)")
plt.legend()
plt.grid()

# Z velocity
plt.subplot(3, 1, 3)
plt.plot(time, vel_no_filter[:, 2], label="Unfiltered Velocity Z")
plt.plot(time, vel_no_filter_filtered[:, 2], label="Filtered Velocity Z")
plt.xlabel("Time (s)")
plt.ylabel("Velocity Z (m/s)")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
